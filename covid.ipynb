```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("country_wise_latest.csv") 
```

```python
print(df.head())
```

```python
print(df.columns)

```

```python
top10_confirmed = df.groupby('Country/Region')['Confirmed'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top10_confirmed.values, y=top10_confirmed.index, palette="viridis")
plt.title("Top 10 Countries by Confirmed Cases")
plt.xlabel("Total Confirmed Cases")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
top5_deaths = df.sort_values(by='Deaths', ascending=False).head(5)

plt.figure(figsize=(8,5))
plt.barh(top5_deaths['Country/Region'], top5_deaths['Deaths'], color='salmon')
plt.title("Top 5 Countries by Deaths")
plt.xlabel("Deaths")
plt.ylabel("Country")
plt.tight_layout()
plt.show()

```

```python
top5_new = df.sort_values(by='New cases', ascending=False).head(5)

plt.figure(figsize=(6,6))
plt.pie(top5_new['New cases'], labels=top5_new['Country/Region'], autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Countries  Share of New Cases")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
top5 = df.sort_values(by='Confirmed', ascending=False).head(5)

plt.figure(figsize=(8,5))
plt.plot(top5['Country/Region'], top5['Deaths'], label='Deaths', marker='o', color='red')
plt.plot(top5['Country/Region'], top5['Recovered'], label='Recovered', marker='s', color='green')
plt.title("Deaths vs Recovered in Top 5 Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.show()
```

```python

india_data = df[df['Country/Region'] == 'India']
```

```python
print(india_data)
```

```python
values = india_data[['Confirmed', 'Active', 'Deaths', 'Recovered']].values[0]
labels = ['Confirmed', 'Active', 'Deaths', 'Recovered']
plt.figure(figsize=(6,4))
plt.bar(labels, values, color=['blue', 'orange', 'red', 'green'])
plt.title("COVID-19 Stats in India")
plt.ylabel("Number of Cases")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'orange', 'red', 'green'])
plt.title("India COVID-19 Case Distribution")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(['New cases', 'New deaths'], [india_data['New cases'].values[0], india_data['New deaths'].values[0]], marker='o')
plt.title("India: New Cases vs New Deaths")
plt.ylabel("Count")
plt.grid(True)
plt.tight_layout()
plt.show()
```
